```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.welcometothejungle.com/fr/jobs?refinementList%5Boffices.country_code%5D%5B%5D=FR&refinementList%5Boffices.state%5D%5B%5D=Ile-de-France&refinementList%5Bcontract_type%5D%5B%5D=apprenticeship&page=1&aroundQuery=%C3%8Ele-de-France%2C%20France&query=data%20engineer"
```

```python
response = requests.get(url)
html_content = response.content
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

```python
print(soup.prettify())
```

```python
# Find all jobs
job_cards = soup.find_all('li')

```

```python
print(job_cards)
```

```python
def get_job_setion_name_info(soup):
    try:
        setion_job_info = soup.find_all("div", class_="sc-1gjh7r6-7 fCBkLE")
        
    except AttributeError:
            setion_job_info = []
            
    return setion_job_info
```

```python
def get_job_setion_type_salary_info(soup):
    try:
        setion_job_info = soup.find_all("div", class_="sc-bXCLTC eFiCOk")
    except AttributeError:
        setion_job_info = []
    return setion_job_info    
```

```python
job_setion_names = get_job_setion_name_info(soup)
job_setion_type_salary = get_job_setion_type_salary_info(soup)
```

```python
def get_job_name(soup):
    try:
        blockJobInfo = soup.find("a")
        jobTitleH4 = blockJobInfo.find("h4", class_="sc-gvZAcH bOVOkk sc-1gjh7r6-1 dakUgn wui-text")
        jobTitleText = jobTitleH4.get_text()
    except AttributeError:
        jobTitleText = ""
    return jobTitleText    
```

```python
print(len(job_setion_names))
print(job_setion_names[0].prettify())
```

```python
print(job_setion_type_salary[0].prettify())
```

```python
def get_job_city(soup):
    try:
        blockAddressInfo = soup.find("div", class_="sc-bXCLTC hZOXdq")
        subBlockAddress = blockAddressInfo.find("div", class_="sc-bXCLTC fYUqim")
        addressBalise = subBlockAddress.find("p", "sc-bXCLTC fYUqim")
        addressText = addressBalise.get_text()
    except AttributeError:
        addressText = ""
    return addressText        
```

```python
def get_job_category(soup):
    try:
        block_category_info = soup.find("div", "sc-bOhtcR bJqGFP")
        span_category_info = block_category_info.find("span")
        category_text = span_category_info.get_text()
    except AttributError:
        category_text = ""
    return category_text     
```

```python
def get_job_salary(soup):
    try:
        block_salary_info = soup.find("div", class_="sc-bOhtcR bJqGFP")
        span_salary_info = block_salary_info.find("span")
        salary_info_text = span_salary_info.get_text()
    except AttributeError:
        salary_info_text = ""
    return salary_info_text
```

```python
def get_type_work(soup):
    try:
        block_type_work = soup.find("div", class_="sc-bOhtcR bJqGFP")
        type_work_info = block_type_work.find("span")
        type_work_text = type_work_info.get_text()
    except AttributeError:
        type_work_text = ""
    return type_work_text
```

```python
job_result = {"Names": [], "Categories": [], "Cities": [], "WorkType": [], "Salaries": []}

for detailinfo in job_setion_type_salary:
    job_result["Categories"].append(get_job_category(detailinfo))
    job_result["WorkType"].append(get_type_work(detailinfo))
    job_result["Salaries"].append(get_job_salary(detailinfo))
    
for mainInfo in job_setion_names:
    job_result["Names"].append(get_job_name(mainInfo))
    job_result["Cities"].append(get_job_city(mainInfo))
    
print(job_result)
```
